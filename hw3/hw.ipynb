```python
from os import listdir

folder = "VCTK-Corpus\\wav48\\p250\\"
wav_files = list(map(lambda f: folder + f, listdir(folder)))

txt_files = [f.replace('wav48', 'txt').replace('.wav', '.txt') for f in wav_files]
whole_list = list(zip(wav_files, txt_files))
k = 0.7
partition = int(len(whole_list) * k)
train_data = whole_list[:partition]
test_data = whole_list[partition:]

with open('train.csv', 'w') as f_train:
    for w, t in train_data:
        f_train.write(w+','+t+'\n')
with open('test.csv', 'w') as f_test:
    for w, t in test_data:
        f_test.write(w+','+t+'\n')
```
